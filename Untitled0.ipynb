```python
import tensorflow as tf
from tensorflow import keras
from keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataIN = []
dataOU = []
with open("learnNNEND.txt") as f:
    dataOU = [[int(i) for i in line.strip()] for line in f.readlines()]
with open("learnNN.txt") as f:
    for line in f:
        dataIN.append([int(x) for x in line.strip().split()])
```

```python
input = np.array(dataIN)
output = np.array(dataOU)
model = keras.Sequential(
    [
     layers.Dense(24, activation='linear'),
     layers.Dense(24, activation='linear'),
     layers.Dense(1, activation='softmax')])
model.compile(optimizer ='adam', loss = 'mean_squared_error', metrics = 'accuracy')
history = model.fit(input, output[:, -1:], batch_size = 12, epochs = 10)
print(history.history)
```
